```python
import requests
```

```python
BASE_URL = 'https://maayanlab.cloud/Enrichr'
```

Modified from https://maayanlab.cloud/Enrichr/help#api

```python
url = f'{BASE_URL}/addList'
genes_str = '\n'.join([
    'PHF14', 'RBM3', 'MSL1', 'PHF21A', 'ARL10', 'INSR', 'JADE2', 'P2RX7',
    'LINC00662', 'CCDC101', 'PPM1B', 'KANSL1L', 'CRYZL1', 'ANAPC16', 'TMCC1',
    'CDH8', 'RBM11', 'CNPY2', 'HSPA1L', 'CUL2', 'PLBD2', 'LARP7', 'TECPR2', 
    'ZNF302', 'CUX1', 'MOB2', 'CYTH2', 'SEC22C', 'EIF4E3', 'ROBO2',
    'ADAMTS9-AS2', 'CXXC1', 'LINC01314', 'ATF7', 'ATP5F1'
])
description = 'Example gene list'
payload = {
    'list': (None, genes_str),
    'description': (None, description)
}

r = requests.post(url, files=payload)
r
```

```python
result = r.json()
result
```

```python
url = f'{BASE_URL}/enrich'
query_string = '?userListId=%s&backgroundType=%s'
user_list_id = result['userListId']
gene_set_library = 'GO_Biological_Process_2021'

params = {
    'userListId': result['userListId'],
    'backgroundType': gene_set_library
}
r = requests.get(url, params=params)
r
```

```python
r.json()
```

```python
keys = ['Rank', 'Term name', 'P-value', 'Z-score', 'Combined score', 'Overlapping genes', 'Adjusted p-value', 'Old p-value', 'Old adjusted p-value']
```

```python
results = []
for ls in r.json()[gene_set_library]:
    results.append({k:v for k, v in zip(keys, ls)})
results
```
